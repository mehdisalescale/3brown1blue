## 4. Visualizing what M does to space

Draw the standard basis (î, ĵ) and a dense grid of points before the transformation, then again after. The grid should be visibly warped.

Sketch on paper first. Predict, then plot.

```python
def make_grid(n: int = 11, span: float = 5.0) -> np.ndarray:
    """Return an (N, 2) array of points on a square grid."""
    xs = np.linspace(-span, span, n)
    ys = np.linspace(-span, span, n)
    xx, yy = np.meshgrid(xs, ys)
    return np.stack([xx.ravel(), yy.ravel()], axis=1)


def plot_grid_and_basis(points: np.ndarray, ax, title: str = "") -> None:
    ax.scatter(points[:, 0], points[:, 1], s=8, alpha=0.6)
    ax.arrow(0, 0, 1, 0, head_width=0.15, length_includes_head=True, color="#3B83BD")
    ax.arrow(0, 0, 0, 1, head_width=0.15, length_includes_head=True, color="#E6B800")
    ax.set_aspect("equal")
    ax.set_xlim(-6, 6)
    ax.set_ylim(-6, 6)
    ax.set_title(title)


grid = make_grid()
transformed = grid @ M.T  # equivalent to applying M to each row-as-column

fig, axes = plt.subplots(1, 2, figsize=(10, 5))
plot_grid_and_basis(grid, axes[0], "Before")
plot_grid_and_basis(transformed, axes[1], "After M = [[2,1],[0,1]] (shear)")
plt.show()
```

## 5. Animate the morph from I to M

Linear interpolation between the identity and `M` over time `t ∈ [0, 1]`:

$$M(t) = (1 - t) \\cdot I + t \\cdot M$$

Watch how the basis vectors and the grid move continuously. Pick three matrices to animate (rotation, shear, projection, reflection — pick what interests you).

```python
I = np.eye(2)


def animate_transform(M_target: np.ndarray, title: str = "", n_frames: int = 60):
    fig, ax = plt.subplots(figsize=(6, 6))
    grid = make_grid()

    def update(frame: int):
        ax.clear()
        t = frame / (n_frames - 1)
        M_t = (1 - t) * I + t * M_target
        transformed = grid @ M_t.T
        plot_grid_and_basis(transformed, ax, f"{title} — t = {t:.2f}")

    anim = FuncAnimation(fig, update, frames=n_frames, interval=40, repeat=False)
    plt.close(fig)
    return HTML(anim.to_jshtml())


theta = np.pi / 2
M_rotate = np.array([[np.cos(theta), -np.sin(theta)], [np.sin(theta), np.cos(theta)]])
animate_transform(M_rotate, "Rotation 90°")
```
